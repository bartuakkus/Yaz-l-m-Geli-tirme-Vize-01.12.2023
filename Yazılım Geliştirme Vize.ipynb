```python
import sys
import pandas as pd

class Kullanici:
    def __init__(self, ad, kullanici_id, mail, sifre):
        self.ad = ad
        self.kullanici_id = kullanici_id
        self.mail = mail
        self.sifre = sifre

    def kayit(self):
        print(f"{self.ad} adlı kullanıcı sisteme kaydedildi.")

    def giris(self):
        print(f"{self.ad} adlı kullanıcı sisteme giriş yaptı.")

    def cikis(self):
        print(f"{self.ad} adlı kullanıcı sistemden çıkış yaptı.")

    def sifre_kontrol(self, girilen_sifre):
        if self.sifre == girilen_sifre:
            print(f"{self.ad} adlı kullanıcının şifresi doğru.")
            return True
        else:
            print(f"{self.ad} adlı kullanıcının şifresi yanlış.")
            return False

class KullaniciSistemi:
    def __init__(self):
        self.kullanicilar = []

    def kullanici_ekle(self, kullanici):
        self.kullanicilar.append(kullanici)

    def kullanici_kaldir(self, kullanici_ad=None, kullanici_id=None):
        for kullanici in self.kullanicilar:
            if kullanici_ad and kullanici.ad == kullanici_ad:
                self.kullanicilar.remove(kullanici)
                print(f"{kullanici_ad} adlı kullanıcı listeden kaldırıldı.")
                break
            elif kullanici_id and kullanici.kullanici_id == kullanici_id:
                self.kullanicilar.remove(kullanici)
                print(f"Kullanıcı ID {kullanici_id} olan kullanıcı listeden kaldırıldı.")
                break
        else:
            print("Kullanıcı bulunamadı.")

    def listele(self):
        for kullanici in self.kullanicilar:
            print(f"Ad: {kullanici.ad}, ID: {kullanici.kullanici_id}, Mail: {kullanici.mail}")

class Ogrenci(Kullanici):
    def __init__(self, ad, kullanici_id, mail, sifre, ogrenci_no):
        super().__init__(ad, kullanici_id, mail, sifre)
        self.ogrenci_no = ogrenci_no

class OgretimGorevlisi(Kullanici):
    def __init__(self, ad, kullanici_id, mail, sifre, bolum):
        super().__init__(ad, kullanici_id, mail, sifre)
        self.bolum = bolum

    def not_ver(self, ogrenci, ders, notu):
        print(f"{self.ad} adlı öğretim görevlisi, {ogrenci.ad} adlı öğrenciye {ders} dersinden not verdi: {notu}")

# Kullanıcı sistemi oluştur
kullanici_sistemi = KullaniciSistemi()

ogrenci1 = Ogrenci("Ahmet", 123, "ahmet@mail.com", "ahmet123", "O12345")
ogretim_gorevlisi1 = OgretimGorevlisi("Prof. Ayşe", 456, "ayse@mail.com", "ayse456", "Bilgisayar Mühendisliği")

# Kullanıcıları sisteme ekliyoruz
kullanici_sistemi.kullanici_ekle(ogrenci1)
kullanici_sistemi.kullanici_ekle(ogretim_gorevlisi1)

while True:
    print("1-Öğrenci Girişi")
    print("2-Öğretim Görevlisi Girişi")
    print("3-Kullanıcı Kaldır")

    secim = input("Öğrenci girişi için 1'i, Öğretim görevlisi girişi için 2'yi, kullanıcı kaldırmak için 3'ü tuşlayınız!")

    if secim == "1":
        ogrenci1.kayit()
        ogrenci1.giris()
        ogrenci1.cikis()
        break
    elif secim == "2":
        ogretim_gorevlisi1.kayit()
        ogretim_gorevlisi1.giris()
        ogretim_gorevlisi1.not_ver(ogrenci1, "Matematik", 90)
        ogretim_gorevlisi1.cikis()
        break
    elif secim == "3":
        kullanici_ad = input("Kaldırmak istediğiniz kullanıcının adını giriniz: ")
        kullanici_sistemi.kullanici_kaldir(kullanici_ad=kullanici_ad)
        break

# Kullanıcıları listele
kullanici_sistemi.listele()
```
